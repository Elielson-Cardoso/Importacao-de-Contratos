```python
nltk.download('punkt')
lista_de_palavras = nltk.tokenize.word_tokenize(textoCRU)
lista_de_palavras
```

```python
lista_de_palavras = [palavra.lower() for palavra in lista_de_palavras]
lista_de_palavras
```

```python
pontuacao = ['(',')',';',':','[',']',',']
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('portuguese')
stop_words
```

```python
keyword = [palavra for palavra in lista_de_palavras if not palavra in stop_words and not palavra in pontuacao]
keyword
```

```python
len(keyword)
```

```python
textocv = " ".join(s for s in keyword)
textocv
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(background_color = '#0f54c9',
                      max_font_size = 150,
                      width = 1280,
                      height = 720,
                      colormap = 'Blues').generate(textocv)


import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (16, 9))
ax.imshow(wordcloud)
ax.set_axis_off()
plt.imshow(wordcloud)
wordcloud.to_file("wordcloud.png")
plt.show()
```
